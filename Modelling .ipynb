```python
lapsed = data[data.STATUS == 'LAPSED' ].copy()
lapsed['freq'] = lapsed.groupby('GRPNUM')['GRPNUM'].transform('count')
lapsed = lapsed.drop_duplicates(subset = 'GRPNUM')
lapsed[lapsed.freq>1]
```

```python
count = 0 ;
one_time_lapsed_grps = []
for i in lapsed.GRPNUM:
    if lapsed[lapsed.GRPNUM == i].freq.any() == 1:
        scan = len(data[data.GRPNUM == i])
        a = data[data.GRPNUM == i].STATUS
        if a.iloc[scan-1] != 'LAPSED':
            count = count + 1
            one_time_lapsed_grps.append(i)
print count   
```

```python
def lapsed_group_treatment(data , GRPNUM , frequency):
    instances = len(data[data.GRPNUM == GRPNUM])
    for i in range(0,instances):
        if frequency.any() == 0:
            break
        else:
            if data[data.GRPNUM == GRPNUM].iloc[i].STATUS == 'LAPSED':
                key = data[data.GRPNUM == GRPNUM].iloc[i].sno
                #print key
                data.iloc[key,2] = 'ACTIVE'
                frequency = frequency - 1
                #print 'success'
```

```python
for GRPNUM in multiple_time_lapsed_grps:
    frequency = lapsed[lapsed.GRPNUM == GRPNUM].freq
    lapsed_group_treatment(data , GRPNUM , frequency)
```

```python
for GRPNUM in one_time_lapsed_grps:
    frequency = lapsed[lapsed.GRPNUM == GRPNUM].freq
    lapsed_group_treatment(data , GRPNUM , frequency)
```
