```python
#!/usr/bin/env python2
# -*- coding: utf-8 -*-
"""
Created on Fri Jun  1 11:31:43 2018

@author: koushik
"""

import pandas as pd
import numpy as np
from sklearn.preprocessing import Normalizer
import matplotlib.pyplot as plt


data=pd.read_csv('/home/koushik/Assignment1/titanic.csv')
y=data.iloc[:,0]
y=y.values.reshape(1,887)
x=data.drop(['Survived','Name','Sex'], axis=1)
x=x.fillna(x.mean())
scaler = Normalizer().fit(x)
x = scaler.transform(x)

w=np.matlib.zeros((5,1))
b=0;
m=len(x)
l=0.5
z=np.dot(w.T,x.T)+b
a=1/(1+np.exp(-z))

def cost(y,z):
    a=1/(1+np.exp(-z))
    return -(np.dot(np.log(a),y.T)+(np.dot(np.log(1-a),(1-y).T)))


def grad_w(w,b):
    a=1/(1+np.exp(-z))
    dz=a-y
    dw=(1/float(m))*(np.dot(x.T,dz.T))
    #w=w-l*dw
    return dw

def grad_b(w,b):
    a=1/(1+np.exp(-z))
    dz=a-y
    db=(1/float(m))*(np.sum(dz))
   # b=b-l*db
    return db

k=[]

for i in range(1,1000):
    db=grad_b(w,b)
    dw=grad_w(w,b)
    w=w-l*dw
    b=b-l*db
    z=np.dot(w.T,x.T)+b
    a=1/(1+np.exp(-z))
    k=np.append(k,cost(y,z))
    print w
    #if(dw[1,1]<0.00001 and dw[1,1]>-0.00001):
     #   break
    
# =============================================================================
# t1=np.arange(0,99,1)    
# plt.plot(t1,k)
# plt.show()
# =============================================================================



```
