```python
def myencode(data,column) : 
    le = preprocessing.LabelEncoder()
    le.fit(data[column])
    dup = le.transform(data[column])
    return dup
```

```python
myencode(data,column='PARENT1')
```

```python
def myencodeAll(data,column) : 
    le = preprocessing.LabelEncoder()
    for col in column:
        le.fit(data[col])
        data[col] = le.transform(data[col])
    return data
```

```python
myencodeAll(data,column=['PARENT1','CAR_TYPE','RED_CAR','REVOKED','URBANICITY'])
```
